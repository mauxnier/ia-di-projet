#### Réalisation du challenge 1

```python
"""
Project: AD4IDS - Anomaly Detection for Intrusion Detection Systems
Subproject: Challenge 1 - HTTPWeb & SSH
Stage: 2- Evaluation
Authors: MONNIER Killian & BAKKARI Ikrame
Date: 01/2024
"""
import json
import pickle
```

Récupère le dataframe depuis le fichier pickle

```python
# with open("data/df_HTTPWeb.pkl", "rb") as f:
with open("data/df_SSH.pkl", "rb") as f:
    df_test = pickle.load(f)

df_test = df_test.drop(columns=["tag_Attack"])
df_test.head()
```

2. Chargement du modèle Random Forest

```python
# Chemin vers le fichier .pkl où le modèle est enregistré
# file_path = "data/random_forest_classifier_HTTPWeb.pkl"
file_path = 'data/random_forest_classifier_SSH.pkl'

# Charger le modèle entraîné à partir d'un fichier
with open(file_path, "rb") as file:
    rf_model = pickle.load(file)
```

3. Prédiction des étiquettes pour les données extraites

```python
predictions = rf_model.predict(
    df_test
)  # Remplacer data_test par les données extraites
probs = rf_model.predict_proba(df_test)  # Probabilités des prédictions
```

4. Création du fichier de résultats

```python
res = {
    "preds": ['Normal' if pred == 0 else 'Attack' for pred in predictions.tolist()],  # Convertit les prédictions en liste pour JSON
    "probs": probs.tolist(),  # Convertit les probabilités en liste
    "names": ["Ikrame BAKKARI", "Killian MONNIER"],
    "method": "Random Forest",
    # "appName": "HTTPWeb",
    "appName": "SSH",
    "version": 1
}
# with open("IADI_IKI_HTTPWeb.res", "w") as f:
with open("IADI_IKI_SSH.res", "w") as f:
    json.dump(res, f)
```
